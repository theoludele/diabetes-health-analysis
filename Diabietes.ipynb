```python
from scipy.stats import chi2_contingency
```

```python
chi2_contingency(crosstab)
```

```python
test_stat, p_value, dof, expected_freq = chi2_contingency(crosstab)
```

```python
p_value
```

```python
def test_association(df, col1, col2):
    level_of_sig = 0.05
    crosstab = pd.crosstab(df[col1], df[col2])
    test = chi2_contingency(crosstab)
    p_value = round(test[1], 2)
    print(f'Test of association between {col1} and {col2}')
    crosstab.plot(kind = 'bar')
    plt.show()
    print(f'P-Value: {p_value}')
    if p_value < level_of_sig:
        print(f'There is an association between {col1} and {col2}')
    else:
        print(f'There is no association between {col1} and {col2}')
        
```

```python
test_association(df, 'Diabetes_Status', 'HighBP')
```

```python
test_association(df, 'Sex', 'Stroke')
```
